```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos desde un archivo CSV
# Montamos la unidad donde se encuentra nuestro archivo de datos
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Leemos el archivo CSV
data = pd.read_csv("/content/drive/MyDrive/CDSMT/IntensidadesTiempoBase_Actualizado.csv", sep=",", encoding="latin1")
```

```python

# Estadísticas descriptivas
stats = data[["IMDL", "IMDS", "IMDD"]].describe()
print("Estadísticas descriptivas:")
print(stats)
```

```python
# Gráfico de barras de medias
means = data[["IMDL", "IMDS", "IMDD"]].mean()
means.plot(kind="bar", color=["blue", "green", "red"])
plt.title("Promedio de Intensidad Media Diaria")
plt.ylabel("Intensidad Media")
plt.xticks(rotation=0)
plt.show()
```

```python
# Boxplots para la distribución
sns.boxplot(data=data[["IMDL", "IMDS", "IMDD"]])
plt.title("Distribución de Intensidad Media Diaria por tipo de día")
plt.ylabel("Intensidad Media")
plt.xticks([0, 1, 2], ["Laborables", "Sábados", "Domingos"])
plt.show()
```

```python
# Tendencias temporales
sns.lineplot(data=data, x="Año", y="IMDL", label="Laborables", marker="o")
sns.lineplot(data=data, x="Año", y="IMDS", label="Sábados", marker="o")
sns.lineplot(data=data, x="Año", y="IMDD", label="Domingos", marker="o")
plt.title("Tendencias de Intensidad Media Diaria por tipo de día")
plt.ylabel("Intensidad Media")
plt.xlabel("Año")
plt.legend()
plt.show()
```
